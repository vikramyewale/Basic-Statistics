```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Q1) Look at the data given below. Plot the data, find the outliers and find out  μ,σ,σ^2

```python
X=[24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00]
Name=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
df=pd.DataFrame(list(zip(X,Name)))
```

```python
df.columns=['X',"Name"]
df.head()
```

```python
plt.figure(figsize=(7,7))
df['X'].plot(kind="pie",autopct="%1.1f%%")
```

```python
sns.boxplot(df.X)
```

```python
print('Mean of X =',df['X'].mean())
print('Median of X =',df['X'].median())
print('Std.dev.  =',df['X'].std())
print('Variance =',df['X'].var())
```

```python
mean=df['X'].mean()
std=df['X'].std()
```

```python
lower_bound=mean-(3*std)
upper_bound=mean+(3*std)
print(lower_bound,',',upper_bound)
```

```python
len(df[(df["X"]<lower_bound) | (df["X"]>upper_bound)])
```
